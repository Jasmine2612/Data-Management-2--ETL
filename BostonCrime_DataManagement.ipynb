```python
##Consistency

df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("WED","WEDNESDAY") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("TUE","TUESDAY") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("FRI","FRIDAY") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("MON","MONDAY") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("THURS","THURSDAY") 

df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("MONDAY","1") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("TUESDAY","2") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("WEDNESDAY","3") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("THURSDAY","4") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("FRIDAY","5") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("SATURDAY","6") 
df["DAY_OF_WEEK"]= df["DAY_OF_WEEK"].replace("SUNDAY","7") 

df["UCR_PART"]= df["UCR_PART"].replace("PART ONE","1") 
df["UCR_PART"]= df["UCR_PART"].replace("PART TWO","2") 
df["UCR_PART"]= df["UCR_PART"].replace("PART THREE","3") 
df["UCR_PART"]= df["UCR_PART"].replace("OTHER","4") 


```

```python
#Conformity
df= df.drop(df[df['OCCURRED_ON_DATE'].str.contains('PM')].index)
df= df.drop(df[df['OCCURRED_ON_DATE'].str.contains('AM')].index)
df['OCCURRED_ON_DATE'] = pd.to_datetime(df['OCCURRED_ON_DATE'])#, format='%d/%m/%Y %H:%M')#.dt.strftime('%m/%d/%Y %H:%M'))
```

```python
#Accurate
#Validity
df['MONTH']= df['MONTH'].replace('ER',np.nan) 
df['MONTH']= df['MONTH'].replace('DSCSC',np.nan) 
#df['MONTH']= df['MONTH'].replace(' ',np.nan) 
df['MONTH']= df['MONTH'].replace('AB',np.nan) 
df['MONTH']= df['MONTH'].replace('AS',np.nan) 
df['MONTH']= df['MONTH'].replace('BH',np.nan) 
df['MONTH']= df['MONTH'].replace('ASASAA',np.nan) 
df['MONTH']= df['MONTH'].replace('CFCS',np.nan) 
df['MONTH']= df['MONTH'].replace('545',np.nan) 
df['MONTH']= df['MONTH'].replace('21',np.nan) 
```

```python
df['MONTH']=df['MONTH'].fillna(df['OCCURRED_ON_DATE'].dt.month)
```

```python
df['DAY_OF_WEEK']=df['DAY_OF_WEEK'].fillna(df['OCCURRED_ON_DATE'].dt.dayofweek)
```

```python
df['HOUR']=df['HOUR'].fillna(df['OCCURRED_ON_DATE'].dt.hour)
```

```python
#street column
df['STREET']=df['STREET'].str.replace(r'\d+','')
df['STREET']=df['STREET'].fillna(np.nan)
```

```python
cursor.execute("CREATE DATABASE BOSTON_CRIME04")
```

```python
cursor.execute("USE BOSTON_CRIME04")
```

```python
cursor.execute("""CREATE TABLE OFFENSE
(OFFENSE_ID int NOT NULL PRIMARY KEY,
OFFENSE_CODE int NULL,
OFFENSE_CODE_GROUP varchar(255) NULL,
OFFENSE_DESCRIPTION varchar(255) NULL,
OFFENSE_DETAILS varchar(255) NULL)""")
```

```python
cursor.execute("""CREATE TABLE OFFENSE_LOCATION
(LOCATION_ID int NOT NULL PRIMARY KEY,
STREET varchar(255) NULL,
LATITUDE float NULL,
LONGITUDE float NULL,
LOCATION varchar(255) NULL)""")
```

```python
cursor.execute("""CREATE TABLE INCIDENT
(INCIDENT_ID int NOT NULL PRIMARY KEY,
INCIDENT_NUMBER varchar(255) NULL,
DISTRICT varchar(255) NULL,
REPORTING_AREA int NULL,
OCCURRED_ON_DATE DATETIME NULL,
YEAR int NULL,
MONTH int NULL,
DAY_OF_WEEK int NULL,
HOUR int NULL,
UCR_PART int NULL,
OFFENSE_ID int NOT NULL,
LOCATION_ID int NOT NULL,
FOREIGN KEY(OFFENSE_ID) REFERENCES OFFENSE(OFFENSE_ID),
FOREIGN KEY(LOCATION_ID) REFERENCES OFFENSE_LOCATION(LOCATION_ID))""")
```

```python
db.commit()
```
